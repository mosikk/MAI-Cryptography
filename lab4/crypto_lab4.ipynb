### Генерация рандомных текстов

Напишем функцию для генерации текстов из рандомных букв и слов.

Для генерации из букв будем использовать только английские буквы нижнего регистра и пробел.  
Для генерации из слов будем использовать все слова английского языка, разделяя их пробелами. Если последнее слово оказалось длиннее, чем нужно, то обрежем его.

```python
import numpy as np
import random
```

```python
def get_random_chars_text(size):
    """
    Generate text with random chars and spaces with given len
    """
    letters = string.ascii_lowercase + ' '

    random_chars_text = np.random.choice(list(letters), size, replace=True)
    random_chars_text = ''.join(random_chars_text)
    assert len(random_chars_text) == size
    return random_chars_text
```

```python
get_random_chars_text(100)
```

```python
with open('english_words.txt') as f:
    english_words = [word.strip() for word in f.readlines()]
```

```python
def get_random_words_text(size):
    """
    Generate text with random words with given len
    """
    
    cur_len = 0
    words_for_text = []
    while cur_len <= size:
        random_word = random.choice(english_words)
        cur_len += len(random_word) + 1
        words_for_text.append(random_word)
    
    result = ' '.join(words_for_text)[:size]
    assert len(result) == size
    return result
```

```python
get_random_words_text(100)
```

### Сравнение текстов

Напишем функцию для сравнения текстов. Будем использовать метрику accuracy

```python
def accuracy(text1, text2):
    assert len(text1) == len(text2)
    matched = 0
    for c1, c2 in zip(text1, text2):
        if c1 == c2:
            matched += 1
    
    return matched / len(text1) * 100
```

#### 1. Сравниваем два нормальных текста

```python
print(f'Точность при сравнении осмысленных текстов = {accuracy(text1, text2)}%')
```

#### 2. Сравниваем нормальный текст с текстом из случайных букв

```python
print(f'Точность при сравнении осмысленного текста и текста из рандомных букв = {accuracy(get_random_chars_text(len(text1)), text2)}%')
```

#### 3. Сравниваем нормальный текст и текст из случайных слов

```python
print(f'Точность при сравнении осмысленного текста и текста из рандомных слов = {accuracy(get_random_words_text(len(text2)), text2)}%')
```

#### 4. Сравниваем два текста из рандомных букв

```python
print(f'Точность при сравнении двух текстов из рандомных букв = {accuracy(get_random_chars_text(len(text1)), get_random_chars_text(len(text2)))}%')
```

#### 5. Сравниваем два текста из рандомных слов

```python
print(f'Точность при сравнении двух текстов из рандомных слов = {accuracy(get_random_words_text(len(text2)), get_random_words_text(len(text2)))}%')
```

### Вывод

Выполнив данную лабораторную работу, я получил следующие результаты:

* точность при сравнении текста из рандомных слов с чем-нибудь = ~6%  
* точность при сравнении текста из рандомных букв с чем-нибудь = ~3.7%

Видим, что при использовании слов (даже рандомных) точность совпадений выше, чем при использовании рандомных букв. В словах буквы связаны друг с другом: в них встречаются частые буквосочетания, частые паттерны итд. В тексте из рандомных букв между символами нет абсолютно никакой связи. Мы можем получить абсурдные сочетания букв и слов, которые маловероятно встретятся в реальном тексте.

Но точность при сравнении текстов из рандомных слов и реальных текстов тоже отличается. Тексты из слов не учитывают связей между этими словами. А реальные тексты, очевидно, учитывают.

При помощи таких оценок мы можем примерно понять, какой текст мы получили (допустим, он зашифрован и мы не можем просто посмотреть на него глазами, или он на незнакомом языке, или он слишком большой и мы не можем изучить его полностью). Мы можем посчитать точности совпадений с реальными текстами и посмотреть на полученное значение. По этому показателю мы сможем понимать, написано ли в тексте что-то осмысленное или это просто набор букв. А если точность околонулевая, то возможно текст вообще на другом языке. Такие проверки используются в некоторых алгоритмах криптографии.
